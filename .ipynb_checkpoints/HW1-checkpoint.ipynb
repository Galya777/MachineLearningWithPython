# Задача 1

В заданието  имате данните отностно сърдечни заболявания, които са взети от [това състезание в kaggle](https://www.kaggle.com/datasets/fedesoriano/heart-failure-prediction).
Данните, които трябва да прочетете, са файлът `hearth.csv`

1.1 Прочетете набора от данни с помощта на `pandas`. (*hint: read_csv*)

1.2 След което разбийте данните на атрибути и целеви атрибут. в случая целевия атрибут е HeartDisease.

1.3 Подгответе категорийните променливи, така че да могат да бъдат обработвани като индикатори. (*hint: get_dummies*)

1.4 Разбийте данните на тренировъчно и тестово множество (като 30% от данните да са в тестовото множество).


```python
import pandas as pd
from sklearn.model_selection import train_test_split

df =pd.read_csv("heart.csv")
print(df.head())

X = df.drop("HeartDisease", axis=1)
y = df["HeartDisease"]

X = pd.get_dummies(X, drop_first=True)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=42
)

print("Размер на тренировъчните данни:", X_train.shape)
print("Размер на тестовите данни:", X_test.shape)
```

# Задача 2

2.1 Напревете логистична регресия, като я натренирате върху тренировъчните данни. Тествайте колко добре се справя върху тренировъчните и тестовите данни, използвайки функцията `.score()`.

2.2 Тествайте как ще се промени резултата като добавите по-голяма регулизация.

2.3 Тествайте как ще се промени резултата като добавите по-малка регулизация. 

2.4 Обеснете как интерпретирате резултатите от екпериментите си.

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(max_iter=1000)

log_reg.fit(X_train, y_train)

train_score = log_reg.score(X_train, y_train)
test_score = log_reg.score(X_test, y_test)

print("Точност върху тренировъчни данни:", train_score)
print("Точност върху тестови данни:", test_score)


log_reg_strong = LogisticRegression(C=0.1, max_iter=1000) # по-голяма регуляризация
log_reg_strong.fit(X_train, y_train)

print("Train score (силна рег.):", log_reg_strong.score(X_train, y_train))
print("Test score  (силна рег.):", log_reg_strong.score(X_test, y_test))


log_reg_weak = LogisticRegression(C=10, max_iter=1000) # по-слаба регуляризация
log_reg_weak.fit(X_train, y_train)

print("Train score (слаба рег.):", log_reg_weak.score(X_train, y_train))
print("Test score  (слаба рег.):", log_reg_weak.score(X_test, y_test))


#Силна (малко C, напр. 0.1)	Моделът е по-ограничен.	Намалява риска от overfitting, 
#понякога влошава точността върху train, но тестът може да стане по-стабилен.
#Слаба (голямо C, напр. 10)	Моделът има по-голяма свобода да се напасне към данните.	
#Често увеличава точността върху train, но може да доведе до overfitting и тестовата точност 
#може да падне.
```

# Задача 3

3.1 Създайте модел за квадратични характеристики (`степен = 2`). Използвайте PolynomialFeatures, за да създадете данните.

3.2 Повторете стъпките от Задача 2 върху квадратичните данни.

3.3 Намерете броя на атрибутите.

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LogisticRegression

poly = PolynomialFeatures(degree=2, include_bias=False)

# Преобразуваме данните
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

log_reg_poly = LogisticRegression(max_iter=2000)
log_reg_poly.fit(X_train_poly, y_train)

print("Точност train (poly):", log_reg_poly.score(X_train_poly, y_train))
print("Точност test  (poly):", log_reg_poly.score(X_test_poly, y_test))

print("Брой атрибути преди:", X_train.shape[1])
print("Брой атрибути след:", X_train_poly.shape[1])


```

# Задача 4

Повторете стъпките от задача 3 като преди това скалирате данните между 0 и 1. Сравнете резултата.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

poly = PolynomialFeatures(degree=2, include_bias=False)

X_train_scaled_poly = poly.fit_transform(X_train_scaled)
X_test_scaled_poly = poly.transform(X_test_scaled)

log_reg_scaled_poly = LogisticRegression(max_iter=2000)
log_reg_scaled_poly.fit(X_train_scaled_poly, y_train)

print("Точност train (scaled + poly):", log_reg_scaled_poly.score(X_train_scaled_poly, y_train))
print("Точност test  (scaled + poly):", log_reg_scaled_poly.score(X_test_scaled_poly, y_test))

print("Брой атрибути преди:", X_train.shape[1])
print("Брой атрибути след:", X_train_poly.shape[1])
```
